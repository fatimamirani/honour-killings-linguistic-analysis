```python
import pandas as pd
import matplotlib.pyplot as plt 

file = 'articles description.xlsx'
df = pd.read_excel(file)

df.head()
```

```python
print(df.columns.tolist())
```

```python
# Counting the AgencyType 
agency_counts = df['AgencyType '].value_counts()

# Plotting
plt.figure(figsize=(8, 5))
agency_counts.plot(kind='bar', color=['#F94144', '#90BE6D', '#577590'])
plt.title('Agency and Voice Representation in Honour Killing Headlines')
plt.xlabel('Agency/Voice Type')
plt.ylabel('Number of Headlines')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
